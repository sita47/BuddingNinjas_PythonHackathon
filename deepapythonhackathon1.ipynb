### 68. Display the data where actual LOS is greater than expected LOS


```python
dfdata = pd.read_excel("HospitalDatabase.xlsx", "ReAdmissionRegistry", index_col = 'PatientID', 
                       parse_dates = ["DischargeDate", "AdmissionDate" ])
dfdata.head()
```

```python
# Check the format of the data
dfdata.info()
```

```python
# # Convert the dates to datetime format

#  We’ll be using the to_datetime() function to convert the dates to datetime
#  objects. The errors='coerce' argument with return NaT (or “not a time”) for any date
#  that cannot be converted. We’ll reassign the order_date and despatch_date columns to datetime
# objects and save them to the original Pandas dataframe columns.
```

```python
dfdata['AdmissionDate'] = pd.to_datetime(dfdata['AdmissionDate'])
dfdata['DischargeDate'] = pd.to_datetime(dfdata['DischargeDate'])
dfdata['ExpectedLOS'] = pd.to_datetime(dfdata['ExpectedLOS'])
```

```python
# We can now check the data types of the columns by running df.info() again. As you can see from the output, the order_date and despatch_date columns are now datetime objects.

dfdata.info()
```

```python
# Calculate the difference between two dates as a timedelta

dfdata['ActualLOS']  = dfdata['DischargeDate'] - dfdata['AdmissionDate'] 
dfdata.head()
```

```python
dfdata['ActualLOS'] = dfdata['ActualLOS'].astype('datetime64[ns]' )
```

```python
dfdata.head()
```

```python
# Select dataframe rows using a condition

dfdata[dfdata['ActualLOS']  > dfdata['ExpectedLOS']]
```

```python
df_merge_2['Date'].apply(lambda x: from_dob_to_age(x))
```
